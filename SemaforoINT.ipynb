```python

# SEMÁFORO INTELIGENTE

# Biblioteca para controlar o tempo
import time


# FUNÇÃO QUE DEFINE O TEMPO DO VERDE

def calcular_tempo_verde(qtd_carros):
    """
    Recebe a quantidade de carros e retorna
    o tempo que o semáforo ficará verde.
    """
    # Se tiver poucos carros
    if qtd_carros <= 5:
        return 10
    # Trânsito médio
    elif qtd_carros <= 15:
        return 20
    # Muitos carros
    else:
        return 30
# Laço de repetição
while True:
    print("\n===== SEMÁFORO INTELIGENTE =====")
    # "sensor" calcula quantos carros tem na via
    qtd_carros = int(input("Quantidade de carros detectados: "))

    # Calcula automaticamente o tempo do verde
    verde = calcular_tempo_verde(qtd_carros)
    # SINAL VERDE
    print("\n VERDE")
    print(f"Tempo liberado: {verde} segundos")
    time.sleep(verde)

    # SINAL AMARELO
    print("\n AMARELO")
    time.sleep(3)
    # SINAL VERMELHO
    print("\n VERMELHO")
    time.sleep(5)

    # Reinicia o ciclo automaticamente
```
