```python
import time

estado = "VERMELHO"
total_carros = 0

while True:
    print("\n--- SEMÁFORO INTELIGENTE ---")
    print("Estado atual:", estado)
    print(f"Carros na via: {total_carros}")
    print("1 - Abrir sinal (Verde)")
    print("2 - Atenção (Amarelo) -> [Aviso: Mudará para Vermelho após 10s]")
    print("3 - Fechar sinal (Vermelho)")
    print("4 - Atualizar quantidade de carros")
    print("5 - Ver status detalhado")
    print("0 - Sair")

    opcao = input("Escolha uma opção: ")

    if opcao == "1":
        estado = "VERDE"
        print("Sinal aberto para os veículos.")

    elif opcao == "2":
        estado = "AMARELO"
        print("Atenção! O sinal mudará para VERMELHO em 10 segundos...")


        for i in range(10, 0, -1):
            print(f"{i}...", end=" ", flush=True)
            time.sleep(1)

        print("\nTempo esgotado!")
        estado = "VERMELHO"
        print("Sinal fechado automaticamente. Veículos devem parar.")

    elif opcao == "3":
        estado = "VERMELHO"
        print("Sinal fechado. Veículos devem parar.")

    elif opcao == "4":
        try:
            total_carros = int(input("Digite a quantidade de carros na via: "))
            print(f"Quantidade de carros atualizada para: {total_carros}")
        except ValueError:
            print("Por favor, digite um número inteiro válido.")

    elif opcao == "5":
        print(f"\n--- STATUS DO SISTEMA ---")
        print(f"Sinal: {estado}")
        print(f"Fluxo atual: {total_carros} carros.")

    elif opcao == "0":
        print("Sistema encerrado.")
        break

    else:
        print("Opção inválida.")
```
