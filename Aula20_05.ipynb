```python
saldo = 1000
historico = []
print(f'Seu saldo é {saldo}')

limite = 850

while True:
  print('Digite 1 para ver saldo  ')
  print('Digite 2 para depositar ')
  print('Digite 3 para sacar ')
  print('Digite 4 para ver o histórico de transferencia ')
  print('Digite 5 para sair')

  escolha = input()

  if escolha == '1':
    print(f'Seu saldo é R$ {saldo}')

  elif escolha == '2':
    deposito = float(input('Quanto deseja depositar'))
    if deposito > 0:
      deposito += saldo
      historico.append(deposito)
      print(f'R${deposito} adicionado, seu saldo atual é de {saldo}')
    else:
      print('Esse valor é invalido')

  elif escolha == '3':
    valor = float(input('Digite o valor que deseja sacar'))
    if valor > limite:
      print(f'!LIMITE DE SAQUE É R${limite}!')
    elif valor > saldo:
      print(f'Saldo insuficiente ')
    elif valor < 0:
      print('Valor invalido')
    else:
      saldo -= valor
      historico.append(valor)
      print(f'Saque de R${valor} feito com sucesso')
  elif escolha == '4':
    print("\n=== HISTÓRICO ===")
    if not historico:
            print("Nenhuma operação realizada.")
    else:
            for operacao in historico:
                print(operacao)

  elif escolha == "5":
        print("Saindo do sistema...")
        break

  else:
        print("Opção inválida!")
```
